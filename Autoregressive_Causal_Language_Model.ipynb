```python
import numpy as np

# -----------------------------
# Random Seed
# -----------------------------
np.random.seed(25)

# -----------------------------
# Sample Token IDs
# (Output from BPE)
# -----------------------------
tokens = np.array([4,8,15,16,23,42,10,6,9,14,18,27,31,12,7])

sequence_length = 4

X = []
Y = []

# -----------------------------
# Create Input-Target Pairs
# -----------------------------
for i in range(len(tokens)-sequence_length):
    X.append(tokens[i:i+sequence_length])
    Y.append(tokens[i+sequence_length])

X = np.array(X)
Y = np.array(Y)

print("Input Sequences")
print(X)

print("\nTarget Tokens")
print(Y)

# -----------------------------
# Embedding Layer
# -----------------------------
vocab_size = 50
embedding_dim = 8

embedding = np.random.randn(vocab_size, embedding_dim)

embedded_input = embedding[X]

print("\nEmbedded Shape :", embedded_input.shape)

# -----------------------------
# Positional Encoding
# -----------------------------
position = np.arange(sequence_length)

pos_encoding = np.random.randn(sequence_length, embedding_dim)

embedded_input = embedded_input + pos_encoding

print("\nPositional Encoding Added")

# -----------------------------
# Causal Mask
# -----------------------------
mask = np.triu(np.ones((sequence_length, sequence_length)), k=1)

print("\nCausal Mask")
print(mask)

# -----------------------------
# Self Attention
# -----------------------------
Q = embedded_input
K = embedded_input
V = embedded_input

scores = np.matmul(Q, K.transpose(0,2,1))

scores = scores / np.sqrt(embedding_dim)

scores = np.where(mask==1, -1e9, scores)

attention = np.exp(scores)
attention = attention / attention.sum(axis=-1, keepdims=True)

context = np.matmul(attention, V)

print("\nContext Shape :", context.shape)

# -----------------------------
# Linear Layer
# -----------------------------
W = np.random.randn(embedding_dim, vocab_size)

logits = np.matmul(context[:, -1, :], W)

print("\nLogits Shape :", logits.shape)

# -----------------------------
# Softmax
# -----------------------------
exp_logits = np.exp(logits - np.max(logits, axis=1, keepdims=True))

probabilities = exp_logits / exp_logits.sum(axis=1, keepdims=True)

print("\nPrediction Probabilities")
print(probabilities)

# -----------------------------
# Predicted Token
# -----------------------------
predicted_tokens = np.argmax(probabilities, axis=1)

print("\nPredicted Next Tokens")
print(predicted_tokens)

# -----------------------------
# Cross Entropy Loss
# -----------------------------
loss = -np.mean(np.log(probabilities[np.arange(len(Y)), Y] + 1e-9))

print("\nLoss :", loss)

# -----------------------------
# Backpropagation
# -----------------------------
print("\nBackpropagation Step (Concept Demonstration)")

# -----------------------------
# Text Generation
# -----------------------------
print("\nGenerated Next Token IDs")

current = list(tokens[:sequence_length])

for i in range(5):

    emb = embedding[np.array(current[-sequence_length:])]
    emb = emb + pos_encoding

    q = emb
    k = emb
    v = emb

    score = np.matmul(q, k.T) / np.sqrt(embedding_dim)

    score = np.where(mask==1, -1e9, score)

    att = np.exp(score)
    att = att / att.sum(axis=-1, keepdims=True)

    ctx = np.matmul(att, v)

    logit = np.matmul(ctx[-1], W)

    prob = np.exp(logit - np.max(logit))
    prob = prob / prob.sum()

    next_token = np.argmax(prob)

    current.append(next_token)

print(current)

print("\nProject Completed Successfully!")
```
