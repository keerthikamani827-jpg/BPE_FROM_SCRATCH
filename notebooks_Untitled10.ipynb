```python
import numpy as np
from collections import defaultdict

print("Libraries imported successfully!")

# -------------------------------
# Corpus
# -------------------------------
corpus = [
    "low", "lower", "lowest",
    "new", "newer",
    "wide", "wider", "widest",
    "playing", "played", "player", "players"
]

print("\nCorpus:")
print(corpus)

# -------------------------------
# Create Vocabulary
# -------------------------------
vocab = {}

for word in corpus:
    chars = tuple(list(word) + ["</w>"])
    vocab[chars] = vocab.get(chars, 0) + 1

print("\nVocabulary:")
for word, freq in vocab.items():
    print(word, ":", freq)

# -------------------------------
# Get Pair Frequencies
# -------------------------------
def get_stats(vocab):
    pairs = defaultdict(int)

    for word, freq in vocab.items():
        for i in range(len(word) - 1):
            pairs[(word[i], word[i + 1])] += freq

    return pairs

# -------------------------------
# Merge Vocabulary
# -------------------------------
def merge_vocab(pair, vocab):
    new_vocab = {}

    for word, freq in vocab.items():
        new_word = []
        i = 0

        while i < len(word):
            if i < len(word)-1 and word[i] == pair[0] and word[i+1] == pair[1]:
                new_word.append(word[i] + word[i+1])
                i += 2
            else:
                new_word.append(word[i])
                i += 1

        new_vocab[tuple(new_word)] = freq

    return new_vocab

# -------------------------------
# Pair Frequencies
# -------------------------------
pairs = get_stats(vocab)

print("\nAdjacent Pair Frequencies:\n")
for pair, freq in pairs.items():
    print(pair, ":", freq)

best = max(pairs, key=pairs.get)

print("\nMost Frequent Pair:")
print(best)

print("\nFrequency:")
print(pairs[best])

# -------------------------------
# One Merge Demo
# -------------------------------
temp_vocab = merge_vocab(best, vocab)

print("\nVocabulary After One Merge:\n")
for word, freq in temp_vocab.items():
    print(word, ":", freq)

# -------------------------------
# Multiple Merges
# -------------------------------
num_merges = 10
merge_rules = []

for i in range(num_merges):
    pairs = get_stats(vocab)

    if not pairs:
        break

    best = max(pairs, key=pairs.get)

    print(f"Merge {i+1}: {best} Frequency = {pairs[best]}")

    merge_rules.append(best)

    vocab = merge_vocab(best, vocab)

# -------------------------------
# Learned Rules
# -------------------------------
print("\nLearned Merge Rules:\n")

for i, rule in enumerate(merge_rules):
    print(f"{i+1}. {rule}")

# -------------------------------
# Final Vocabulary
# -------------------------------
print("\nFinal Vocabulary:\n")

for word, freq in vocab.items():
    print(word, ":", freq)

# -------------------------------
# Vocabulary Tokens
# -------------------------------
tokens = set()

for word in vocab:
    for token in word:
        tokens.add(token)

print("\nVocabulary Tokens:\n")

for token in sorted(tokens):
    print(token)

print("\nVocabulary Size:", len(tokens))

# -------------------------------
# Encode Function
# -------------------------------
def encode(word, vocab):
    tokens = list(word) + ["</w>"]

    changed = True

    while changed:
        changed = False

        for pair in merge_rules:
            i = 0
            new = []

            while i < len(tokens):
                if i < len(tokens)-1 and tokens[i] == pair[0] and tokens[i+1] == pair[1]:
                    new.append(tokens[i] + tokens[i+1])
                    i += 2
                    changed = True
                else:
                    new.append(tokens[i])
                    i += 1

            tokens = new

    return tokens

# -------------------------------
# Decode Function
# -------------------------------
def decode(tokens):
    return "".join(tokens).replace("</w>", "")

# -------------------------------
# Test Words
# -------------------------------
test_words = [
    "lowest",
    "lower",
    "widest",
    "players",
    "newer",
    "playing"
]

for word in test_words:
    encoded = encode(word, vocab)
    decoded = decode(encoded)

    print("\nOriginal :", word)
    print("Encoded :", encoded)
    print("Decoded :", decoded)
    print("-"*40)
```
